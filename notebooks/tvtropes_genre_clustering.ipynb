```python
import sys
sys.path.append('..')   
import pandas as pd
from pathlib import Path

df = pd.read_csv('../data/tvtropes/lit_goodreads_match.csv')
df.head()
# genre_filtered = df[df['genre'].isin(genres_tropes)]
df = df[df['Trope'].isin(genres_tropes)]
df.drop(columns=['Unnamed: 0'], inplace=True)
df.to_csv('lit_goodreads_genre_filtered.csv', index=False)
df.groupby('Trope').count()
```

```python

# unique titles
title_ids = df['title_id'].unique()
print(len(title_ids))

BOOKS_DIR = Path("../data/books")
books_downloaded = [f.stem for f in BOOKS_DIR.iterdir() if f.is_file()]
books_downloaded = [book for book in books_downloaded if book in title_ids]
print(f'books downloaded: {len(books_downloaded)}/{len(title_ids)}')

titles_to_download = set(title_ids.tolist()) - set(books_downloaded)
titles_to_download = list(titles_to_download)
print(f'titles to download: {len(titles_to_download)}')
```
